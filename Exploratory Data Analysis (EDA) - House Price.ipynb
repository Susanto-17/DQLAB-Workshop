### 2.a. Menampilkan Missing Value 

```python
def cek_data_hilang(df):
    jumlah_nan = df.isna().sum().reset_index()
    jumlah_nan.columns = ['Kolom', 'Jumlah_NaN']
    jumlah_nan["Persentase"] = jumlah_nan.Jumlah_NaN/len(df)
    return jumlah_nan.sort_values("Persentase", ascending = False)
tabel_data_hilang = cek_data_hilang(df)
```

```python
tabel_data_hilang
```

### 2.b. Plotting sum null values dengan bar chart

### 2.c. handling missing value drop more than 0.7 

```python
hilang = tabel_data_hilang[tabel_data_hilang.Persentase < 0.7] #code Persentase dari code diatasnya
hilang
```

```python
to_fix_col = hilang[hilang.Persentase > 0].Kolom.tolist()
```

```python
# drop
df2 = df.drop(tabel_data_hilang[tabel_data_hilang.Persentase > 0.7].Kolom.tolist(), 1)
to_fix = df2[df2[to_fix_col].describe().columns.tolist()]
to_fix.skew()
```

```python
# Function Visualisasi Facegrid
def viz(df, types):
    num = df
    f = pd.melt(num, value_vars = num)
    g = sns.FacetGrid(f, col="variable", col_wrap = 3, sharex = False, sharey = False, size = 5)
    g = g.map(types, "value")
    plt.show
    return (g)

import warnings
warnings.filterwarnings("ignore")
viz(to_fix, sns.distplot) # LotFrontage with median , GarageYrBlt mean , and MasVnrArea with median
```

```python
# LotFrontage with median , GarageYrBlt mean , and MasVnrArea with median
df2[['MasVnrArea','LotFrontage']] = df2[['MasVnrArea','LotFrontage']].fillna(df2[['MasVnrArea','LotFrontage']].median())
df2['GarageYrBlt'] = df2['GarageYrBlt'].fillna(df2['GarageYrBlt'].mean())
```

```python
# categorical fixing with modus
col_to_cat_to_fix = df2[to_fix_col].describe(include = 'object').columns.tolist()
df2[col_to_cat_to_fix] = df2[col_to_cat_to_fix].fillna(df2[col_to_cat_to_fix].mode())
```

### 4.D. Outliers Report 

```python
# outliers report
def outliers(s):
    iqr = (np.quantile(s, 0.75))-(np.quantile(s, 0.25))
    upper_bound = np.quantile(s, 0.75)+(1.5*iqr)
    lower_bound = np.quantile(s, 0.25)-(1.5*iqr)
    f = []
    for i in s:
        if i > upper_bound:
            f.append(i)
        elif i < lower_bound:
            f.append(i)
    sums = len(f)
    pros = len(f)/len(s)*100
    d = {'IQR':iqr,
         'Upper Bound':upper_bound,
        'Lower Bound':lower_bound,
        'Sum outliers': sums,'percentage outliers':pros}
    d = pd.DataFrame(d.items(),columns = ['sub','values'])
    return(d)
```

### 5.b. Korelasi plot dengan heatmap 

```python
def corr_plot(df2, metode):
  plt.figure(figsize=(30,15))
  triangle = np.triu(np.ones_like(df2.corr(method = metode)))
  heatmap = sns.heatmap(df2.corr(), vmin=-1, vmax=1, annot=True, mask = triangle)
  heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':15}, pad=12)
  return heatmap
```

### 5.e. Identifikasi multicolinearity dengan variance inflation factore

    1. Apa itu Multicolenearity
      Multikolinearitas adalah sebuah situasi yang menunjukkan adanya korelasi atau hubungan kuat antara dua variabel bebas atau lebih dalam sebuah model regresi berganda.

    2. Mengapa perlu mencari Multicol?
    Menurut Ghozali (2016) pada pengujian multikolinearitas bertujuan untuk mengetahui apakah model regresi ditemukan adanya korelasi antar variabel independent atau variable bebas. Efek dari multikolinearitas ini adalah menyebabkan tingginya variabel pada sampel. 

    3. standard VIF (Variance Inflation Factor) : suatu alat untuk mengukur multicolenearity pada konteks multiple regresi variable. standard VID di terima apa bila < 10. Apa bila > 10 berarti variabel itu dikatakan multicolenearitas

image.png

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):
    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return(vif)
```

```python
# encoding categorical to see multicol
for i in df2_categoric:
  df2_categoric[i] = df2_categoric[i].astype('category').cat.codes
```

```python
# join numeric and categorical again
df2_categoric = df2_categoric.reset_index()
df2_numeric = df2_numeric.reset_index()
df3 = pd.merge(df2_numeric, df2_categoric, how = 'inner', on = 'Id')
```

```python
X = df3.drop('SalePrice',1)
vif = calc_vif(X)
vif = vif.sort_values('VIF', ascending=False)
vif['VIF'] = round(vif.VIF)
vif[vif.VIF < 10]
```

```python
# def barplot comparison
def barplot_value(x, y, title):
  chart = sns.barplot(x, y, ci = None)
  chart.set_title(title)
  for p in chart.patches:
        chart.annotate("%.0f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
            ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
            textcoords='offset points')
  return chart      
```

### c. Jawaban no. 4 : berapa jumlah rumah yang dijual dengan kondisi penjualan normal?

```python
sale_conditions = df[['SalePrice', 'SaleCondition']].groupby('SaleCondition').agg('mean')
sale_conditions
```

```python
sale_conditions['Jumlah_sale_condition_Type'] = df['SaleCondition'].value_counts()
sale_conditions['prosentase_KondisiPenjualan'] = df['SaleCondition'].value_counts()/len(df)*100
sale_conditions = sale_conditions.reset_index()
sale_conditions 
```

##### Custome Pie Chart

```python
# Creating explode data
explode = (0.2, 0.8, 0.65, 0.5, 0.0, 0.0)
 
# Creating color parameters
colors = ( "orange", "cyan", "brown",
          "grey", "indigo", "beige")
 
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
 
# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} %)".format(pct, absolute)
 
# Creating plot
fig, ax = plt.subplots(figsize =(18, 18))
wedges, texts, autotexts = ax.pie(sale_conditions.prosentase_KondisiPenjualan,
                                  autopct = lambda pct: func(pct, sale_conditions.prosentase_KondisiPenjualan),
                                  explode = explode,
                                  labels = sale_conditions.SaleCondition,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))
 
# Adding legend
ax.legend(wedges, sale_conditions.SaleCondition,
          title ="Cars",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
 
plt.setp(autotexts, size = 12, weight ="bold")
ax.set_title("Customizing pie chart")
 
# show plot
plt.show()
```
